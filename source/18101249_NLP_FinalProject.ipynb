```python
import random

_initial_rules = [
  ["ㄱ", "┐", "￢", "ワ", "ケ", "⺄"],
  ["ㄲ", "ワワ", "┐┐", "￢￢", "⺄⺄"],
	["ㄴ", "Ŋ", "じ", "└"],
  ["ㄷ", "㈂", "⊂", "で"],
  ["ㄸ", "⊂⊂", "でで"],
  ["ㄹ", "乙", "ⓡ"],
  ["ㅁ", "ロ", "▧"],
  ["ㅂ", "ㅂ"],
  ["ㅃ", "ㅃ"],
  ["ㅅ", "ん", "∧", "人", "ベ", "ハ"],
  ["ㅆ", "んん", "∧∧"],
  ["ㅇ", "Ω", "Ø", "Θ", "ø", "㉧"],
  ["ㅈ", "ズ", "㈈", "ス"],
  ["ㅉ", "ズズ", "㈈㈈", "スス"],
  ["ㅊ", "大"],
  ["ㅋ", "ヲ"],
  ["ㅌ", "㉫", "モ", "Ε"],
  ["ㅍ", "ㅍ"],
  ["ㅎ", "ㆅ", "ð", "あ"]
]

# ㅏ ㅐ ㅑ ㅒ ㅓ ㅔ ㅕ ㅖ ㅗ ㅘ ㅙ ㅚ ㅛ ㅜ ㅝ ㅞ ㅟ ㅠ ㅡ ㅢ ㅣ
_middle_rules = [
  ["ㅏ", "┞", "「", "├", "ド"],            
  ["ㅐ", "├│", "Ｈ", "Й", "け", "ℍ"], 
  ["ㅑ", "ギ", "F"],
  ["ㅒ", "ギ|", "Ħ"],               
  ["ㅓ", "ナ", "┧"],
  ["ㅔ", "┧┃"],
  ["ㅕ", "=|"],
  ["ㅖ", "=il"],             
  ["ㅗ", "⑤"],
  ["ㅘ", "ㅗr"],
  ["ㅙ", "ㅗH"],
  ["ㅚ", "ㅗi"],
  ["ㅛ", "ㅛ"],
  ["ㅜ", "u"],
  ["ㅝ", "ㅝ"],
  ["ㅞ", "ㅞ"],
  ["ㅟ", "ㅜi"],
  ["ㅠ", "ⓤ"],
  ["ㅡ", "ㅡ"],
  ["ㅢ", "ㅡ┃", "ㅡI"],
  ["ㅣ", "ノ", "┃", "ı", "！", "∥", "1"],
]

# 중성과 종성을 결합하여 치환할 수 있는 리스트
# 0번 인덱스 : 중성, 1번 인덱스 : 종성
_middle_final_rules = [
  ["ㅏ", "ㅇ", "Б", "6"],            
  ["ㅏ", "ㄴ", "ど", "Ł"],
  ["ㅣ", "ㄴ", "i∽し"],
  ["ㅕ", "ㅇ", "ぎ"]
]

_rules= [
  ["...", "…"],
	["녕", "じギ∽㉧"],
	["들", "Ðμ㉣"],
	["모", "母"],
	["데", "㈂ㅓl", "Ðё"], 
	["!", "¡"],
	["우", "♀"],
	["님", "늼"],
	["왜", "ⓦⓗⓨ"],
	["하", "㈛", "㉻", "あド"],
	[".", "。"],
	["고", "ヱ"],
	["자", "子"],
	["면", "ロぎ"],
	["~", "∽"],
	["이", "ⓔ", "②", "oI"],
	["쁜", "ㅹ"],
	["소", "小"],
	[",", "、"],
	["?", "¿"],
	["그", "ュ"],
  ["마", "㈒"],
  ["조", "죠"],
  ["오", "⑤"],
  ["금", "金"],
  ["카", "㉸"],
  ["붕", "ㅸ"],
  ["쁘", "ㅹ"],
  ["긔", "그ノ"],
  ["셈", "섬l"],
  ["리", "引"],
  ["별", "☆", "★"],
  ["요", "Yo"],
  ["어", "Ю"]
]

def word_rules(text):
  for rule in _rules : 
    randIdx = random.randrange(1, len(rule))
    text = text.replace(rule[0], rule[randIdx])

  return text
```

```python
class Hangeul:
    element_query = {
        # 초성
        'initial' : [ "ㄱ", "ㄲ", "ㄴ", "ㄷ", "ㄸ", "ㄹ", "ㅁ", "ㅂ", "ㅃ",
                "ㅅ", "ㅆ", "ㅇ", "ㅈ", "ㅉ","ㅊ", "ㅋ", "ㅌ", "ㅍ", "ㅎ" ],

        # 중성
        'middle' : [ "ㅏ", "ㅐ", "ㅑ", "ㅒ", "ㅓ", "ㅔ", "ㅕ", "ㅖ", "ㅗ",
                 "ㅘ", "ㅙ", "ㅚ", "ㅛ", "ㅜ", "ㅝ", "ㅞ", "ㅟ", "ㅠ", "ㅡ", "ㅢ", "ㅣ" ],

        # 종성
        'final' : [ "", "ㄱ", "ㄲ", "ㄳ", "ㄴ", "ㄵ", "ㄶ", "ㄷ", "ㄹ", "ㄺ",
                 "ㄻ", "ㄼ", "ㄽ", "ㄾ", "ㄿ", "ㅀ", "ㅁ", "ㅂ","ㅄ", "ㅅ", "ㅆ",
                 "ㅇ", "ㅈ", "ㅊ", "ㅋ", "ㅌ","ㅍ", "ㅎ" ]
    }

    # 생성자
    def __init__(self, letter_or_element):
        if type(letter_or_element) == str:
            self.letter = letter_or_element
            self.element = self.separate(letter_or_element)
        elif type(letter_or_element) == list:
            self.letter = self.combine(letter_or_element)
            self.element = letter_or_element
        else:
            self.letter = ""
            self.element = ["","",""]        

    def __str__(self):        
        return self.letter

    def separate(self, letter):
        x = ord(letter) - ord('가')
        fin_ord = x % 28
        neu_ord = ((x - fin_ord) // 28) % 21
        ini_ord = (((x - fin_ord) // 28)- neu_ord) // 21

        initial = self.element_query['initial'][ini_ord]
        middle = self.element_query['middle'][neu_ord]
        final = self.element_query['final'][fin_ord]
        return [initial, middle, final]

    def combine(self, element):
        ini_ord = self.element_query['initial'].index(element[0])
        neu_ord = self.element_query['middle'].index(element[1])
        fin_ord = self.element_query['final'].index(element[2])

        hangeul_ord = (((ini_ord * 21) + neu_ord) * 28) + fin_ord

        return chr(hangeul_ord + ord('가'))

    # 변환 함수
    def divide_rules(self):
        temp = self
        temp_str = ""

        # 초성 변환
        for initR in _initial_rules : 
            initIdx = random.randrange(1, len(initR))
            temp.element[0] = temp.element[0].replace(initR[0], initR[initIdx])

        # 중성 + 종성의 조합으로 한 번에 바꿀 수 있는 문자가 있을 경우
        is_changed = False
        for midNfinal in _middle_final_rules :
            if midNfinal[0] == temp.element[1] and midNfinal[1] == temp.element[2] :
                midNfinalIdx = random.randrange(2, len(midNfinal))
                temp.element[1] = midNfinal[midNfinalIdx]
                temp.element[2] = ""
                is_changed = True
      
        if is_changed == False :
            # 중성 변환
            for middleR in _middle_rules :
                middleIdx = random.randrange(1, len(middleR))
                temp.element[1] = temp.element[1].replace(middleR[0], middleR[middleIdx])
            # 종성 변환
            for initR in _initial_rules : 
                finalIdx = random.randrange(1, len(initR))
                temp.element[0] = temp.element[0].replace(initR[0], initR[finalIdx])

        temp_str += temp.element[0]
        temp_str += temp.element[1]
        temp_str += temp.element[2]

        return temp_str

# 문장에 대해 한꺼번에 변환하도록
def split_rules(text):
  encrypted = ""
  for letter in text:
    if ord('가') <= ord(letter) <= ord('힣'):
        encrypted += Hangeul(letter).divide_rules()
        #encrypted += Hangeul(letter).divide_rules().letter
    else:
        encrypted += letter
  return encrypted

strText = "안녕하세요! 저랑 자연어처리 공부하실래요? 대박 너무 어렵습니다."
strText2 = "영현.. 잠이 와..?"

# word_rules(통으로 구현되는 경우) -> split_rules(쪼개서 구현되는 경우) 순서로 함수를 적용합니다

strText = word_rules(strText)
strText = split_rules(strText)

strText2 = word_rules(strText2)
strText2 = split_rules(strText2)

print(strText)
print(strText2)
```
